**1. Introducción del problema**

Las redes sociales son plataformas digitales que permiten el intercambio de ideas, intereses u opiniones entre diversos usuarios. Su naturaleza interactiva las han convertido en una herramienta indispensable para las grandes empresas, puesto que resultan efectivas al momento de crear conexión entre una marca y su público objetivo. Es así que estas plataformas se han posicionado como el principal canal publicitario a nivel mundial. Asimismo, son fundamentales para conocer y comprender las opiniones o percepciones del público sobre una campaña publicitaria. Las empresas pueden rastrear, en tiempo real, menciones, comentarios y conversaciones relacionadas, y así saber cómo el público está respondiendo a la campaña. Este tipo de análisis les permite mejorar sus estrategias de marketing para abordar las preocupaciones del cliente y mejorar la percepción de su marca.  

En 2022, Balenciaga lanzó una campaña publicitaria muy polémica, pues presentaba a niños y niñas junto a disfraces bondage. **2.1. Objetivo del proyecto**

El objetivo de este proyecto es elaborar un código para analizar los sentimientos que los usuarios de X manifestaron respecto a la Campaña Navideña 2022 de la marca Balenciaga, a través del uso del hashtag "cancelBalenciaga". En ese sentido, los tweets serán clasificados como positivos, negativos o neutros para proporcionarnos una idea general de cómo fue percibida la campaña publicitaria. **3. Resultados del análisis**

A continuación, presentamos los hallazgos que obtuvimos gracias a la información proporcionada por los gráficos realizados y algunas interpretaciones de a qué se deben estos.

En primer lugar, observamos que los meses en los cuales se usó más el hashtag "#cancelBalenciaga" fueron diciembre 2022 y enero 2023; posteriormente, la cantidad de tweets disminuyó enormemente. Esto resulta coherente al tratarse del periodo de lanzamiento y de máxima exposición de la campaña, por lo que sugiere un mayor interés de discutir y/o compartir opiniones durante este periodo inicial.

En segundo lugar, identificamos que las palabras más utilizadas eran términos con connotaciones fuertes, tales como "pedofilia", "satanismo" o "explotación infantil". Esto sugiere que las imágenes mostradas en la campaña contribuyeron a construir una percepción negativa y muy dañina para la reputación de la marca e incluso llevar a la idea de que la marca está involucrada en actividades ilegales, inmorales y repulsivas. Por consiguiente, la imagen pública, la credibilidad y la confianza de los consumidores de la marca pueden verse enormemente afectados.

En tercer lugar, identificamos una asociación entre Balenciaga y la celebridad Kim Kardashian, puesto que muchos de los comentarios también estaban dirigidos a criticar la colaboración entre este personaje y la marca en cuestión. Algunos se refieren a ella como "pedo supporter", lo que sugiere que esa percepción negativa creada alrededor de la marca, puede trasladarse a cualquiera que trabaje en conjunto con ella.

En cuarto y último lugar, al llevar a cabo nuestro análisis de sentimientos, descubrimos que casi el 50% de los comentarios se clasificaron como neutros, lo que podría indicar una falta de sentimientos fuertes que pueden categorizarse como negativos o positivos. Sin embargo, más del 50% de los comentarios fueron clasificados como negativos, superando ligeramente al grupo de neutros. Esto sugiere una predominancia de sentimientos desfavorables hacia la campaña navideña 2022 de la marca Balenciaga, como indignación y/o repudio, que manifestaron los usuarios de X.

```python
# Importando librerías
import pandas as pd
import matplotlib.pyplot as plt
import re

# Importando Base de Datos Excel
datos = pd.read_excel('base de datos.xlsx')

# Limpiando los tweets
def limpiar_tweets(text):
  text = re.sub(r'http\S+', '', text)
  text = re.sub(r'@[\w_]+', '', text)
  text = re.sub(r'#\w+', '', text)
  return text.strip()

datos['tweets_limpios'] = datos['text'].apply(limpiar_tweets)

# Seleccionando los diez tweets más vistos
tweets_10 = datos.sort_values(by='views', ascending=False).head(10)

# Creando variables para el gráfico
texto = tweets_10['tweets_limpios']
likes = tweets_10['likes']

# Graficando
plt.figure(figsize=(12, 10))
barras = plt.barh(texto, likes, color='skyblue')
plt.xticks(rotation=0)

for bar in barras:
  formato = bar.get_width()
  plt.text(formato, bar.get_y() + bar.get_height()/2, int(formato), va='center')

plt.xlabel('Cantidad de Likes')
plt.ylabel('Tweets')
plt.title('Cantidad de Likes de los 10 Tweets Más Vistos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Importando librerías
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords

# Importando Base de Datos Excel
datos = pd.read_excel('base de datos.xlsx')

# Creando listas para analizar sentimientos
sentimiento_positivo = ['support Balenciaga', 'stand Balenciaga', 'love Balenciaga', 'like Balenciaga', 'amazing', 'good', 'great', 'wonderful', 'lovely']
sentimiento_negativo = ['pedophilia', 'pedo', 'pedo supporter', 'pedophile supporter', 'CancelKimKardashian', 'Kim', 'cancelkardashians', 'satanic', 'fake', 'pedo brand', 'exploitation', 'exploiting', 'children exploitation',
                        'scandal', 'sick', 'kids', 'don t forget', 'never forget', 'cancel', 'cancelled', 'shame', 'business', 'evil', 'mad']

# Limpiando y procesando los tweets
datos['text'] = datos['text'].apply(lambda x: re.sub(r'http\S+', '', x))

def limpiar_tweets(tweet):
  tweet = re.sub(r'[^A-Za-z\s]', '', tweet)
  tweet = tweet.lower()
  return tweet

datos['tweets_limpios'] = datos['text'].apply(limpiar_tweets)

# Analizando los sentimientos de los tweets
def analizar_sentimientos(tweet):
  palabras = tweet.split()
  score = 0
  for palabra in palabras:
    if palabra in sentimiento_positivo:
      score += 1
    elif palabra in sentimiento_negativo:
      score -= 1
  return score

datos['sentimiento'] = datos['tweets_limpios'].apply(analizar_sentimientos)

# Contando el número de tweets positivos, neutrales y negativos
contar_sentimientos = datos['sentimiento'].apply(lambda x: 'Positivo' if x > 0 else ('Negativo' if x < 0 else 'Neutral')).value_counts()

# Graficando
plt.figure(figsize=(8, 6))
plt.pie(contar_sentimientos, labels=contar_sentimientos.index, autopct='%1.1f%%', startangle=140, colors=['red', 'yellow', 'green'])
plt.title('Distribución de Sentimientos de los Tweets')
plt.axis('equal')
plt.show()
```
